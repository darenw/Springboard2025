Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great!

```python
def ratio_for_borough(bname: str, first_year=1998, last_year=2018, verbose=False):
    # df is global. look, don't touch; don't need "global" keyword
    # makes sense to use year_grouped, but i can't get it to work right now.
    bdf = df[ df["Borough"]==bname ]
    early = bdf[bdf["Year"]==first_year]
    late  = bdf[bdf["Year"]==last_year]
    first_price = float( early["avg_price"].mean() )
    last_price = float( late["avg_price"].mean() )
    #print(f"\nBDF  {bdf}\n\nEARLY {early}\n\nLATE {late}")
    if verbose:
       print(f"{bname}")
       print(f"   avg price {first_year}: {first_price}")
       print(f"             {last_year}:  {last_price}")
    return last_price/first_price

#test:
r = ratio_for_borough("Ealing", verbose=True)
print(f"Price ratio for Ealing: {r:.2f}")
```

```python
list_of_boroughs = df["Borough"].unique()
#print(list_of_boroughs)

all_ratios = {}
for bor in list_of_boroughs:
    r = ratio_for_borough(bor)
    all_ratios[bor] = r

#print(all_ratios)
sorted_ratios = sorted(all_ratios.items(),  key=lambda kv: (-kv[1],kv[0]))
print(type(sorted_ratios))
#print("\n", sorted_ratios)

# Should have first and last years to consider in vars somewhere several cells prior to this
# but that'snot the reality at the moment. 
print("\nHighest few price ratios from {1998} to {2018}")
for i in range(6):
    print(sorted_ratios[i])
```
